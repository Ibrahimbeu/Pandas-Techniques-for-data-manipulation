```python
# Applying StandardScaling to columns by hand 
minim= df.age.min()
maxim= df.age.max()
df["age"]= df.apply(func={"age":lambda x: (x-minim)/(maxim-minim) })
display(df[['age']].head())
display(df[['age']].tail())
```

```python
def age_f(x):
    if x<10:
        return 't1'
    elif x>10 and x<20:
        return 't2'
    elif x>20 and x<30:
        return 't3'
    elif x>30 and x<40:
        return "t4"
    elif x>40 and x<50:
        return "t5"
    elif x>50 and x<60:
        return "t6"
    elif x>60 :
        return "t7"
```

```python
df['age_type']= df['age'].apply(age_f)
```

```python
df.groupby(by="age_type")\
.agg({"age":np.mean,"pclass":np.median})\
.groupby("pclass")\
.agg({"age":np.mean})
```
